```python
url = 'http://data.ex.co.kr/openapi/trafficapi/nationalTrafficVolumn?'
serviceKey='key=6844121548&'
type = 'type=xml&'
#exDivCode='exDivCode=00'
exDivCode='exDivCode=01'


```

```python
def parse():
    try:
        carType = list.find('carType').get_text()
        exDivCode = list.find('exDivCode').get_text()
        sumDate = list.find('sumDate').get_text()
        tcsType = list.find('tcsType').get_text()
        trafficVolumn = list.find('trafficVolumn').get_text()
        
        return{
            "차종":carType,
            "구분소기준":exDivCode,
            "집계날짜":sumDate,
            "하이패스/일반구분":tcsType,
            "수량":trafficVolumn
        }
    except AttributeError as e:
            return{
            "차종":None,
            "구분소기준":None,
            "집계날짜":None,
            "하이패스/일반구분":None,
            "수량":None
        }
```

##### 일자별로 각각 다 받아오기 위해 for문을 돌림

```python
row = []
for date in coronaDate:
    result = requests.get(url+serviceKey+type+'sumDate={}&'.format(date)+exDivCode)
    soup = bs(result.text,'lxml-xml')
    lists = soup.find_all('list')
    for list in lists:
        row.append(parse())
    
    
    
```
