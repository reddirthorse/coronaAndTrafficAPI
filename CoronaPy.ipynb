```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#접속 url
url = "http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?"
serviceKey = "serviceKey=rbAs0dbairVckbSl70DYRWMbrbNLNyztrtuuINYAh8Z7OgP7OpUOof6E3hIBWF4wXf8yGYVTObv9oTUkA2ffSA%3D%3D"
numOfRows="&numOfRows=10"
StartCreateDt = "&startCreateDt=20190101"
endCreateDt = "&endCreateDt=20211113"
```

```python
def parse():
    try:
        STATE_DT = item.find("stateDt").get_text()
        DECIDE_CNT = item.find("decideCnt").get_text()
        DEATH_CNT = item.find("deathCnt").get_text()

        
        return {
            "기준일":STATE_DT,
            "확진자 수":DECIDE_CNT,
            "사망자 수":DEATH_CNT,
   
        }
    
    except AttributeError as e:
        
        return {
            "기준일":None,
            "확진자 수":None,
            "사망자 수":None,

        }
```

```python
result = requests.get(url+serviceKey+"&pageNo=1"+numOfRows+StartCreateDt+endCreateDt)
soup = bs(result.text,'lxml-xml')
items = soup.find_all("item")
```

```python
row = []
for item in items:
    row.append(parse())
```
